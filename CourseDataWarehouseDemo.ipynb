<a id = "9875"> <h2> 2. Finance Department </h2> </a>

```python
%%sql

SELECT DISTINCT InstructorID, Name, ProgramName
FROM CourseData.COURSE_MEETINGS
   JOIN CourseData.COURSE_OFFERINGS USING (CourseOfferingID)
    JOIN PROGRAMS_DIM ON (PROGRAMS_DIM.ProgramCode = SUBSTR(CourseData.COURSE_OFFERINGS.CatalogID,1,INSTR(CourseData.COURSE_OFFERINGS.CatalogID," ")-1))
    LEFT JOIN INSTRUCTORS_DIM ON (INSTRUCTORS_DIM.Name = CourseData.COURSE_OFFERINGS.PrimaryInstructorID)

WHERE Term = 'Fall2016' AND ProgramName = 'Finance'
;
```

<a id = "9876"> <h2> 3. Where Is Biology Taught on Thursdays? </h2> </a>

```python
%%sql

SELECT DISTINCT Term, ProgramName, Location, Day
FROM CourseData.COURSE_MEETINGS
   JOIN CourseData.COURSE_OFFERINGS USING (CourseOfferingID)
    JOIN PROGRAMS_DIM ON (PROGRAMS_DIM.ProgramCode = SUBSTR(CourseData.COURSE_OFFERINGS.CatalogID,1,INSTR(CourseData.COURSE_OFFERINGS.CatalogID," ")-1))
    LEFT JOIN INSTRUCTORS_DIM ON (INSTRUCTORS_DIM.Name = CourseData.COURSE_OFFERINGS.PrimaryInstructorID)
    LEFT JOIN LOCATION_DIM USING (LocationID)

WHERE ProgramName = 'Biology' AND Term = 'Fall2017' AND Day = 'R'
LIMIT 10;
```

<a id = "9877"> <h2> 4. Who Taught in the Lecture Halls in Spring 2018? </h2> </a>

```python
%%sql

SELECT DISTINCT Term, Name, ProgramName, Location, Day 
FROM CourseData.COURSE_MEETINGS
   JOIN CourseData.COURSE_OFFERINGS USING (CourseOfferingID)
    JOIN PROGRAMS_DIM ON (PROGRAMS_DIM.ProgramCode = SUBSTR(CourseData.COURSE_OFFERINGS.CatalogID,1,INSTR(CourseData.COURSE_OFFERINGS.CatalogID," ")-1))
    LEFT JOIN INSTRUCTORS_DIM ON (INSTRUCTORS_DIM.Name = CourseData.COURSE_OFFERINGS.PrimaryInstructorID)
    LEFT JOIN LOCATION_DIM USING (LocationID)

WHERE Location = 'DSB 112' AND Term = 'Spring2018'
ORDER BY Name
;
```

<a id = "9878"> <h2> 5. When and Where was Accounting Taught in Spring 2019? </h2> </a>

```python
%%sql

SELECT DISTINCT Term, Name, ProgramName, Location, Day 
FROM CourseData.COURSE_MEETINGS
   JOIN CourseData.COURSE_OFFERINGS USING (CourseOfferingID)
    JOIN PROGRAMS_DIM ON (PROGRAMS_DIM.ProgramCode = SUBSTR(CourseData.COURSE_OFFERINGS.CatalogID,1,INSTR(CourseData.COURSE_OFFERINGS.CatalogID," ")-1))
    LEFT JOIN INSTRUCTORS_DIM ON (INSTRUCTORS_DIM.Name = CourseData.COURSE_OFFERINGS.PrimaryInstructorID)
    LEFT JOIN LOCATION_DIM USING (LocationID)

WHERE Term = 'Spring2019' AND ProgramName = 'Accounting'
ORDER BY Location
LIMIT 10;
```
